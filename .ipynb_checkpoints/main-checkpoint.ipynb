```python
import matplotlib.pyplot as plt
import itertools
import numpy as np
import datetime, os
import tensorflow as tf
import tensorflow.keras.layers as layers
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

```python
from keras.utils.data_utils import validate_file
from keras.preprocessing.image import ImageDataGenerator

# Global settings
BATCH_SIZE = 32
IMG_SIZE = (160, 160)

# Training dataset generator - augments images as they are retrieved.
train_datagen = ImageDataGenerator(preprocessing_function=lambda x: x/255.0,
                                   rotation_range=20,
                                   zoom_range=.2,
                                   shear_range=20,
                                   horizontal_flip=True,
                                   width_shift_range=(-10, 10),
                                   height_shift_range=(-15, 15),
                                   brightness_range=(0.8,1.2),
                                   )

# Training and testing generators - no image augmentation.
test_datagen = ImageDataGenerator(preprocessing_function=lambda x: x/255.0)
valid_datagen = ImageDataGenerator(preprocessing_function=lambda x: x/255.0)

# Iterable data flow objects
train = train_datagen.flow_from_directory('split_dataset/train',
                                    batch_size=32, seed=1337, 
                                    target_size=IMG_SIZE, color_mode='rgb',
                                    class_mode='binary')

valid = valid_datagen.flow_from_directory('split_dataset/val',
                                    batch_size=32, seed=1337, 
                                    target_size=IMG_SIZE, color_mode='rgb',
                                    class_mode='binary')

test = test_datagen.flow_from_directory('split_dataset/test',
                                    batch_size=32, seed=1337, 
                                    target_size=IMG_SIZE, color_mode='rgb',
                                    class_mode='binary')
```

```python
class MacroConv2D(layers.Layer):
    """ Similar to Conv2D but the kernel is a moving window over a larger weight matrix. """
    def __init__(self, filters, window_size, stride=1, padding='VALID', **kwargs):
        self.filters = filters
        self.window_size = window_size
        self.stride = stride
        self.padding = padding

        super(MacroConv2D, self).__init__(**kwargs)

    def build(self, input_shape):
        self.kernel = self.add_weight(name='kernel', 
                                      shape=(1,) + input_shape[1:-1] + (input_shape[-1], self.filters),
                                      initializer='uniform',
                                      trainable=True)
        
        self.bias = self.add_weight(name='bias', 
                                      shape=(self.filters,),
                                      initializer='uniform',
                                      trainable=True)

        super(MacroConv2D, self).build(input_shape[:-1] + (self.filters,))

    def call(self, inputs):
        inputs = tf.expand_dims(inputs, axis=-1)    # Make shapes work
        hadamard_product = inputs * self.kernel     # 
        hadamard_product = tf.math.reduce_sum(hadamard_product, axis=3, keepdims=False)

        pooled = tf.nn.avg_pool2d(hadamard_product, self.window_size, self.stride, self.padding) * np.prod(self.window_size)

        return pooled + self.bias
    
    def get_config(self):
        """ Return hyperparameters. """
        conf_dict = {"filters": self.filters,
                    "window_size": self.window_size,
                    "stride": self.stride,
                    "padding": self.padding}
        return conf_dict
```

```python
# Build model using Keras functional API.
input_image = layers.Input(shape=(160, 160, 3))
x = layers.Conv2D(16, (3,3))(input_image)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D((2,2))(x)

x = layers.Conv2D(12, (3,3))(x)
x = layers.Activation('relu')(x)
x = layers.MaxPooling2D((2,2))(x)

x = layers.Conv2D(8, (3,3))(x)
x = layers.Activation('relu')(x)

x = layers.Flatten()(x)
output_prediction = layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(input_image, output_prediction)
```

```python
# EXPERIMENT ZONE
# Hyperparameters needed for experimentation
learning_rates = (0.1, 0.01, 0.001,)
batch_sizes = (8, 16, 32)
epoch = 500

# Save initial weights
model.save_weights('tf_models/init_weights')
i = 0
for learning_rate, batch_size in itertools.product(learning_rates, batch_sizes):        
    # Tensorboard setup
    tb_name = '_'.join(map(str, ['conv2d', 'lr', learning_rate, 'bs', batch_size]))
    logdir = f"logs/{tb_name}__{datetime.datetime.now().strftime('%Y%m%d-%H%M%S')}"
    tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)
    train = train_datagen.flow_from_directory('split_dataset/train',
                                    batch_size=batch_size, seed=1337, 
                                    target_size=IMG_SIZE, color_mode='rgb',
                                    class_mode='binary')
    
    # Compile model and reset weights
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    model.load_weights('tf_models/init_weights')
    
    # Train model and save
    model.fit(train, epochs=epoch, validation_data=valid, callbacks=[tensorboard_callback])
    model_name = 'tf_models/' + '_'.join(map(str, ['model', 'lr', learning_rate, 'bs', batch_size, 'ep', epoch]))
    model.save(model_name)
```

```python
# Load model and test on test dataset
model = tf.keras.models.load_model('tf_models/thinner_lr_0.001_bs_16_ep_500')
loss, accuracy = model.evaluate(test)
```
